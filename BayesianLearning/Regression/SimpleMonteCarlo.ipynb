```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tqdm import tqdm
```

```python
def f(x):
    return 0.5*x**2 + 0.7*x
```

```python
x = np.arange(-2, 2, 0.01)
plt.plot(x, f(x))
```

```python
def p():
    return np.random.normal(loc = 0.5)
```

# 式
$$ f(x) = 0.5 * x^2 + 0.7 * x $$
$$ p(z) = \mathcal{N}(z | 0.5, 1) $$

```python
N = list(range(1,10000, 10))
E = []
for n in tqdm(N):
    res = 0
    for _ in range(n):
        res += f(p())
    ex = res / n
    E.append(ex)

```

```python
plt.plot(N, E)
plt.grid()
plt.show()
```
