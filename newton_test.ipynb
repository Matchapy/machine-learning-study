```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = np.arange(0, 100, 0.01)
```

```python
sin_y = np.exp(T) * np.sin(T)
sin_y2 = [0]
for t in T[:-1]:
    sin_y2.append(sin_y2[-1] + np.exp(t)*(np.sin(t) + np.cos(t))*0.01)
sin_y2 = np.array(sin_y2)
```

```python
# plt.plot(T, sin_y)
# plt.plot(T, sin_y2)
plt.plot(T, sin_y2-sin_y)
plt.show()
```
